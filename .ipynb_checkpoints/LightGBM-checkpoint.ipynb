```python
import lightgbm as lgb
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv('/content/SVMtrain.csv')
```

```python
data.drop(['Sex'],axis=1,inplace=True)
```

```python
data
```

```python
data.info()
```

```python
# define input and output feature
x = data.drop(['Embarked','PassengerId'],axis=1)
y = data.Embarked
```

```python
data.Embarked.value_counts()
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
model = lgb.LGBMClassifier(random_state=42)
model.fit(x_train,y_train,eval_set=[(x_test,y_test),(x_train,y_train)],
          verbose=20)
```

```python
print('Training accuracy {:.4f}'.format(model.score(x_train,y_train)))
print('Testing accuracy {:.4f}'.format(model.score(x_test,y_test)))
```

```python
lgb.plot_importance(model)
```

```python
lgb.plot_tree(model,figsize=(30,40))
```

```python
lgb.plot_metric(model)
```

```python
metrics.plot_confusion_matrix(model,x_test,y_test,cmap='Blues_r')
```

```python
print(metrics.classification_report(y_test,model.predict(x_test)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_1 = GradientBoostingClassifier().fit(x_train,y_train)
```

```python
print(metrics.classification_report(y_test,model_1.predict(x_test)))
```
